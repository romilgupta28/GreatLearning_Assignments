### K - Means Clustering

### Assign a dummy array called distortion

You can later use this array to plot the elbow plot

```python
distortion = []
```

### Calculating distortion - C distance

Iterating values of k from 1 to 10 fit K means model
Using c distance - Get the measure for Sum of squares error.

**Here, logically K-Means attempts to minimize distortion defined by the the sum of the squared distances between each observation and its closest centroid.**

```python
from sklearn.cluster import KMeans
from scipy.spatial import distance

for k in range(1,10):
    model = KMeans(n_clusters=k).fit(interest_df_z)
    distortion.append(sum(np.min(distance.cdist(interest_df_z, model.cluster_centers_, 'euclidean'), axis=1))/ interest_df_z.shape[0])
```

### Plotting Elbow/ Scree Plot

Use Matplotlib to plot the scree plot - Note: Scree plot plots distortion vs the no of clusters

```python
plt.plot(range(1,10), distortion, 'go-')
```

### Find out the optimal value of K

3

### Using optimal value of K - Cluster the data. 
Note: Since the data has more than 2 dimension we cannot visualize the data. As an alternative, we can observe the centroids and note how they are distributed across different dimensions

```python
kmeans = KMeans(n_clusters=3, n_init = 15, random_state=10)
```

```python
kmeans.fit(interest_df_z)
```

You can use kmeans.cluster\_centers\_ function to pull the centroid information from the instance

```python
centroids = kmeans.cluster_centers_
```

### Store the centroids in a dataframe with column names from the original dataset given 

Hint: Use pd.Dataframe function 

```python
centroid_df = pd.DataFrame(centroids, columns = interest_df_z.columns)
```

## Hierarchical Clustering 

For Hierarchical clustering, we will create datasets using multivariate normal distribution to visually observe how the clusters are formed at the end

a = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]], size=[100,])
b = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], size=[50,])
c = np.random.multivariate_normal([10, 20], [[3, 1], [1, 4]], size=[100,])

```python
np.random.seed(101)  # for repeatability of this dataset
a = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]], size=[100,])
b = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], size=[50,])
c = np.random.multivariate_normal([10, 20], [[3, 1], [1, 4]], size=[100,])
```

### Combine all three arrays a,b,c into a dataframe

```python
X = np.concatenate((a, b, c), axis=0)
print(X.shape)  # 250 samples with 2 dimensions
```

### Use scatter matrix to print all the 3 distributions

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

### Find out the linkage matrix

Use ward as linkage metric and distance as Eucledian

```python
#### generate the linkage matrix
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(X, 'ward', metric='euclidean')
Z.shape
```

```python
a
```

```python
Z[:]
```

### Plot the dendrogram for the consolidated dataframe

```python
plt.figure(figsize=(25, 10))
dendrogram(Z)
plt.show()
```

### Recreate the dendrogram for last 12 merged clusters 

Hint: Use truncate_mode='lastp' attribute in dendrogram function to arrive at dendrogram 

```python
dendrogram(
    Z,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=3,  # show only the last p merged clusters
)
plt.show()
```

### From the truncated dendrogram, find out the optimal distance between clusters which u want to use an input for clustering data

```python
max_d = 50
```

### Using this distance measure and fcluster function to cluster the data into 3 different groups

```python
from scipy.cluster.hierarchy import fcluster
clusters = fcluster(Z, max_d, criterion='distance')
clusters
```

### Use matplotlib to visually observe the clusters in 2D space 

```python
#### plt.figure(figsize=(10, 8))
plt.scatter(X[:,0], X[:,1], c=clusters)  # plot points with cluster dependent colors
plt.show()
```

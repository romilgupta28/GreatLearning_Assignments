```python
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
list_to_labelencoding = ['account_check_status','credit_history','purpose','savings','present_emp_since',
                        'personal_status_sex','other_debtors','property','other_installment_plans','housing',
                        'job','telephone','foreign_worker']
for colname in list_to_labelencoding:
    le = preprocessing.LabelEncoder()
    random_new_data[colname] = le.fit_transform(random_new_data[colname])

random_new_data.head()
    
```

### Q3 Split Train/Test data 70:30 ratio( 1 Marks)
##### Hint:from sklearn.model_selection import train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
X=random_new_data.drop(labels=['default'],axis='columns')
y=random_new_data['default']
test_size=0.30
seed =1 # Random numbmer seeding for reapeatability of the code
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
```

### Q4 Build Random Forest Model( 1 Marks)
#### Hint:from sklearn.ensemble import RandomForestClassifier using n_jobs=2,n_estimators=500,criterion="entropy",random_state=9999

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfcl = RandomForestClassifier(n_estimators=500,criterion="entropy",random_state=9999,n_jobs=2)
rfcl = rfcl.fit(X_train,y_train)
```

```python
test_pred = rfcl.predict(X_test)
rfcl.score(X_test,y_test)
```

### Q5 Calculate Confusion Matrix and Accuracy score (1 Marks)
##### Hint: Use confusion_matrix and accuracy_score

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_test, test_pred))
print(metrics.confusion_matrix(y_test, test_pred))
```

### Bootstrapping ( Bonus)
##### Given a dataset of size n, a bootstrap sample is created by sampling n instances uniformly from the data (with/without replacement)
##### Create a model with each bootstrap sample and validate it with the test set
##### Final result is calculated by averaging the accuracy of models

```python
# Number of iterations for bootstrapping
bootstrap_iteration = 10
accuracy = []
```

```python
from sklearn.utils import resample
from sklearn.metrics import accuracy_score

for i in range(bootstrap_iteration):
    X_, y_ = resample(X_train, y_train)
    rfm.fit(X_, y_)
    y_pred = rfm.predict(X_test)
    
    acc = accuracy_score(y_pred, y_test)
    accuracy.append(acc)
```

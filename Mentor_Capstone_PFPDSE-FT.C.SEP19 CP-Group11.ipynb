```python
# fill missing values of boolean columns with "false"

listings_summary['host_has_profile_pic'].fillna('f',inplace=True)
listings_summary['host_identity_verified'].fillna('f',inplace=True)

# convert t/f to 1/0

listings_summary['is_location_exact'] = listings_summary['is_location_exact'].map({'f':0,'t':1})
listings_summary['host_is_superhost'] = listings_summary['host_is_superhost'].map({'f':0,'t':1})
listings_summary['instant_bookable'] = listings_summary['instant_bookable'].map({'f':0,'t':1})
listings_summary['require_guest_profile_picture'] = listings_summary['require_guest_profile_picture'].map({'f':0,'t':1})
listings_summary['requires_license'] = listings_summary['requires_license'].map({'f':0,'t':1})
listings_summary['host_has_profile_pic'] = listings_summary['host_has_profile_pic'].map({'f':0,'t':1})
listings_summary['host_identity_verified'] = listings_summary['host_identity_verified'].map({'f':0,'t':1})
listings_summary['require_guest_phone_verification'] = listings_summary['require_guest_phone_verification'].map({'f':0,'t':1})
```

```python
# replace $ with comma so that columns become numeric
listings_summary['price'] = listings_summary['price'].str.replace('$', '').str.replace(',', '').astype(float)
listings_summary['extra_people'] = listings_summary['extra_people'].str.replace('$', '').str.replace(',', '').astype(float)
listings_summary['cleaning_fee'] = listings_summary['cleaning_fee'].str.replace('$', '').str.replace(',', '').astype(float)
listings_summary['security_deposit'] = listings_summary['security_deposit'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
# Split "amenities" to multiple columns based on their availability and then drop it

listings_summary['Laptop_friendly_workspace'] = listings_summary['amenities'].str.contains('Laptop friendly workspace')
listings_summary['TV'] = listings_summary['amenities'].str.contains('TV')
listings_summary['Family_kid_friendly'] = listings_summary['amenities'].str.contains('Family/kid friendly')
listings_summary['Host_greets_you'] = listings_summary['amenities'].str.contains('Host greets you')
listings_summary['Smoking_allowed'] = listings_summary['amenities'].str.contains('Smoking allowed')
listings_summary['Hot_water'] = listings_summary['amenities'].str.contains('Hot water')
listings_summary['Fridge'] = listings_summary['amenities'].str.contains('Refrigerator')
listings_summary['Wifi'] = listings_summary['amenities'].str.contains('Wifi')

listings_summary['No_of_amentities'] = listings_summary['amenities'].apply(lambda x:len(x.split(',')))
listings_summary.drop('amenities',axis=1,inplace=True)
```

```python
# treatment of some categorical columns

category = ['neighbourhood_cleansed','neighbourhood_group_cleansed','property_type','room_type','bed_type','cancellation_policy']

for i in category:
    listings_summary[i] = listings_summary[i].astype('category')
    listings_summary[i] = listings_summary[i].cat.codes
```

```python
# fill NA values of reviews_per_month with median

listings_summary['reviews_per_month'] = listings_summary['reviews_per_month'].fillna(listings_summary['reviews_per_month'].median())
```

```python
# fill missing value of bedrooms with 1
listings_summary['bedrooms'].fillna(1.0, inplace=True)
```

```python
# fill NA values of text columns with a common text

listings_summary['description'].fillna('No description available',inplace=True)
listings_summary['name'].fillna('No name available',inplace=True)
listings_summary['zipcode'].fillna('00000',inplace=True)
listings_summary['host_location'].fillna('Missing location',inplace=True)
```

```python
# Create a function to calculate distance from the center of Berlin, and add a column to the data frame

from math import sin, cos, sqrt, atan2, radians

def haversine_distance_central(row):
    berlin_lat,berlin_long = radians(52.5200), radians(13.4050)
    R = 6373.0
    long = radians(row['longitude'])
    lat = radians(row['latitude'])
    
    dlon = long - berlin_long
    dlat = lat - berlin_lat
    a = sin(dlat / 2)**2 + cos(lat) * cos(berlin_lat) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

listings_summary['distance'] = listings_summary.apply(haversine_distance_central,axis=1)

# drop latitude and longitude columns

listings_summary=listings_summary.drop('latitude',axis=1)
listings_summary=listings_summary.drop('longitude',axis=1)
```
